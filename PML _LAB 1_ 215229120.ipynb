## Word cloud

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
text = "Statistics are like bikinis. What they reveal is suggestive, but what they conceal is vital"
wordcloud = WordCloud(width = 400, height =80, random_state=1, background_color='white', collocations=False, stopwords = STOPWORDS).generate(text)
plt.figure(figsize=(30, 20))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

## read and play video

```python
import time
import cv2
```

```python
cap= cv2.VideoCapture('Wild_flowers_in_sunset_2.mp4')


fps= int(cap.get(cv2.CAP_PROP_FPS))

print("This is the fps ", fps)

if cap.isOpened() == False:
    print("Error File Not Found")

while cap.isOpened():
    ret,frame= cap.read()

    if ret == True:

        time.sleep(1/fps)

        cv2.imshow('frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    else:
        break


cap.release()
cv2.destroyAllWindows()
```

## parser

```python
import parser
print("Program to demonstrate parser module in Python")
print("\n")
exp = "4 + 9"
print("The given expression for parsing is as follows:")
print(exp)
print("\n")
print("Parsing of given expression results as: ")
st = parser.expr(exp)
print(st)
print("\n")
print("The parsed object is converted to the code object")
code = st.compile()
print(code)
print("\n")
print("The evaluated result of the given expression is as follows:")
res = eval(code)
print(res)
```
